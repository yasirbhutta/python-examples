```python
import pandas as pd
import numpy as np
```

Importing dataset

The dataset is a filtered and modified version of the [Laptop Price Prediction using specifications](https://www.kaggle.com/datasets/arnabchaki/laptop-price-prediction?resource=download)  dataset, available under the Database Contents License (DbCL) v1.0 on the Kaggle website.

Load the dataset to a pandas dataframe named 'df'

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv'

df = pd.read_csv(url,header=None)
df.head()
```

Add headers to the dataframe

```python
headers = ["Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"]

df.columns = headers
df.head(10)
```

Replace '?' with 'NaN'

```python
df.replace('?',np.nan,inplace=True)
df.head()
```

Print the data types of the dataframe columns

```python
df.dtypes
```

Print the statistical description of the dataset, including that of 'object' data types.

```python
print(df.describe(include='all'))
```
